## CategorÃ­as y comparaciÃ³n de datos

Â¿CuÃ¡l es la media de notas(GPA) por cada especialidad?

```python
avg_gpa_by_major = student_data.groupby(by='Major').GPA.mean().to_frame().sort_values(by='GPA')
plt.figure(figsize=(10,6))

sns.barplot(data=avg_gpa_by_major, x='Major', y='GPA')
```

```python
print(avg_gpa_by_major)
```

Â¿QuiÃ©n dedica mÃ¡s tiempo a estudiar? Â¿Chicos o chicas?

```python
# Checking the mean of both male and female attendance
mean_study_hours_by_gender = student_data.groupby('Gender')['StudyHoursPerWeek'].mean()
print(mean_study_hours_by_gender)
```

```python
sns.barplot(data=student_data, x='Gender', y='StudyHoursPerWeek')
```

Â¿CuÃ¡l es la relaciÃ³n entre attendancerate y parttimejob?

```python
attendance_by_job = student_data.groupby('PartTimeJob')['AttendanceRate'].mean()
print(attendance_by_job)
```

```python
sns.boxplot(x='PartTimeJob', y='AttendanceRate', data=student_data)
plt.title('Tasa de asistencia segÃºn tener un trabajo a tiempo parcial')
plt.show()
```

Â¿Afectan el tener un partime job a las notas?

```python
gpa_by_job = student_data.groupby('PartTimeJob')['GPA'].mean()
print(gpa_by_job)
```

```python
sns.boxplot(x='PartTimeJob', y='GPA', data=student_data)
plt.title('GPA segÃºn tener un trabajo a tiempo parcial')
plt.show()
```
